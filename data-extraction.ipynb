```python
import pandas as pd
books = pd.read_csv(f"{path}/books.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar = False, cmap = "Blues", ax = ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np

books['missing_description'] = np.where(books['description'].isna(), 1, 0 )
books['age_of_book'] = 2025 - books['published_year']
```

```python
#check whether there is some sort of weird relation going on in between pgs, age, description and rating

columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = 'spearman')

sns.set_theme(style="white")
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot = True, cmap = "Blues", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

As there are no value closer to 1 or -1 this indicates that there is no strong correlation

Now i want only those books in which description, number of pages, average rating and published year is available

```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

One problem we found was that there are multiple categories which will make things difficult for ML model to classify books into different categories

Now we want to filter out books with limited description because there will be no point to recommend such books

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

Now we need to merge title as well as subtitle because in some cases subtitle are missing and doing that will help our model to be not overtrained

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(
        book_missing_25_words["subtitle"].isna(),  # Condition: Check if subtitle is NaN
        book_missing_25_words["title"],           # If True: Use only title
        book_missing_25_words[["title", "subtitle"]]  # If False: Concatenate title and subtitle
        .astype(str)                             # Convert to string
        .agg(": ".join, axis=1)                   # Concatenate with ": "
    )
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
import pandas as pd
print(pd.__version__)

(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
